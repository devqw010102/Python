```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family = 'Malgun Gothic')
filename = './../../data/kbo.csv'
```

```python
kbo = pd.read_csv(filename, encoding='cp949')
kbo
```

```python
xdata = kbo.loc[:, ['AVG']]
ydata = kbo.loc[:, ['HR']]

plt.figure()
plt.plot(xdata, ydata, marker = 'o', color = 'brown', linestyle = 'None')
plt.xlabel('타율')
plt.ylabel('홈런')
plt.title('산점도 그래프')
plt.grid(True)
```
