```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family='Malgun Gothic')
cnt, PNG, UNDERBAR = 0, '.png', '_'
CHART_NAME = 'brokenLineExam_연습_문제'
filename = './../../data/연령별_실업율.csv'
```

```python
data = pd.read_csv(filename, encoding='cp949', index_col='연도')

YEAR = ['2001년', '2002년', '2003년', '2004년', '2005년', '2006년', '2007년', '2008년', '2009년', '2010년']
AGEG = ['20대', '30대', '40대', '50대', '60세이상']

chartdata = data.loc[YEAR, AGEG]
chartdata = chartdata.T
```

```python
chartdata.plot(figsize = (10, 6), legend = True, marker = 'o', rot = 0)

# xlabel, ylabel, title

plt.xlabel('연령대')
plt.ylabel('연도명')
plt.title('연령대별 연도명 꺽은 선')
```
