```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from Book.chap03.pieChartExam import horizontalalignment
```

```python
plt.rc('font', family = 'Malgun Gothic')
filename = './../../data/final_exam.csv'

data = pd.read_csv(filename, encoding='cp949')
data
```

```python
chartdata = data['korean']
chartdata
```

```python
index = data['names']
index
```

```python
mycolor = ['b', 'g', 'r', 'c', 'm']
mycolor
```

```python
plt.figure()
plt.bar(index, chartdata, color = mycolor, alpha = 0.7)

, xlabel = '학생 이름', ylabel = '점수', title = '국어 점수'
plt.xlabel('학생 이름')
plt.ylabel('점수')
plt.title('국어 점수')

meanval = chartdata.mean()

average = '평균 : %d점' % meanval
plt.axhline(meanval, color = 'r', linewidth = 1, linestyle = '--')
plt.text(-0.5, meanval + 0.5, average, horizontalalignment = 'left')

# plt.rc('font', size = 7)
for idx in range(chartdata.size) :
    value = format(chartdata.iloc[idx], ',') + '점'
    plt.text(idx, chartdata.iloc[idx] + 1, value, horizontalalignment = 'center')
```
